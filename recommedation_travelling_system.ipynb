# Membangun dan Melatih Model Embedding untuk Rekomendasi Tempat Wisata Berdasarkan Kategori

## Penjelasan input :

1. **Import Pustaka:**

   * `pandas` (untuk manipulasi data),
   * `tensorflow` (untuk membangun model pembelajaran mesin),
   * `numpy` (untuk perhitungan numerik),
   * `matplotlib.pyplot` (untuk visualisasi grafik),
   * `sklearn.metrics` (untuk evaluasi model).

2. **Pembersihan dan Persiapan Data:**

   * **`df.dropna(subset=['Description', 'Rating', 'Category'])`**: Menghapus baris dengan nilai yang hilang di kolom **`Description`**, **`Rating`**, atau **`Category`**.
   * **`df.drop_duplicates(subset=['Place_Id', 'Description'])`**: Menghapus duplikasi tempat wisata berdasarkan **`Place_Id`** dan **`Description`**.
   * **`df.reset_index(drop=True)`**: Mereset indeks DataFrame setelah pembersihan data.

3. **Membangun Model Embedding dengan LSTM:**

   * Kode ini mendefinisikan model **`TextEmbeddingModel`** dengan **LSTM** (Long Short-Term Memory) yang dilengkapi dengan **regularisasi L2** dan **dropout** untuk mencegah overfitting.
   * Model ini menerima input **embedding** teks untuk deskripsi tempat wisata dan mengeluarkan representasi vektor dari teks tersebut.

4. **Tokenisasi dan Padding Deskripsi:**

   * **`Tokenizer`** digunakan untuk mengubah deskripsi teks menjadi urutan angka berdasarkan kata yang paling sering muncul (dalam hal ini, 10.000 kata yang paling umum).
   * **`pad_sequences`** memastikan panjang urutan tetap konsisten (100 kata per deskripsi).

5. **Membangun dan Melatih Model:**

   * **Model** dikompilasi menggunakan **Adam optimizer** dan **loss function MSE** (Mean Squared Error), serta metrik **MAE** (Mean Absolute Error) dan **RMSE** (Root Mean Squared Error).
   * **Early stopping** digunakan untuk menghentikan pelatihan lebih awal jika model tidak menunjukkan perbaikan dalam metrik validasi dalam beberapa epoch.
   * Data dibagi secara otomatis menjadi **80% untuk pelatihan** dan **20% untuk validasi** dengan menggunakan `validation_split`.

6. **Representasi Vektor Deskripsi:**

   * Fungsi **`get_description_embedding`** digunakan untuk mendapatkan representasi vektor untuk deskripsi tempat wisata setelah model dilatih.

7. **Cosine Similarity:**

   * Fungsi **`cosine_similarity_tf`** menghitung kesamaan antar dua vektor deskripsi tempat menggunakan **cosine similarity**, yang mengukur seberapa mirip dua vektor dalam ruang vektorial.
   * **`calculate_similarity_matrix`** menghitung matriks kesamaan untuk semua tempat dalam dataset, membandingkan setiap tempat dengan tempat lainnya.

8. **Rekomendasi Berdasarkan Kategori:**

   * Fungsi **`rekomendasi_top10_kategori`** mengambil input kategori dari pengguna dan memberikan rekomendasi 10 tempat wisata teratas berdasarkan **rating** dan **kesamaan deskripsi** tempat dengan tempat lain di kategori yang sama.
   * Rekomendasi dilakukan dengan menghitung **rata-rata kesamaan** untuk setiap tempat dalam kategori dan mengurutkannya berdasarkan **rating** dan **kesamaan**.

9. **Contoh Penggunaan:**

   * **Kategori yang diminta** adalah **"Cagar Alam"**, dan hasil rekomendasi 10 tempat wisata dengan rating tertinggi dan kesamaan tertinggi dihitung dan ditampilkan.`

Ringkasan:

* Kode ini membangun model pembelajaran mendalam untuk menghasilkan **representasi vektor** deskripsi tempat wisata menggunakan **LSTM**.
* Setelah model dilatih, **cosine similarity** digunakan untuk menghitung kesamaan antar tempat wisata, dan akhirnya digunakan untuk memberikan **rekomendasi tempat wisata** terbaik berdasarkan kategori yang diminta oleh pengguna.


```python
import pandas as pd
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras import metrics
from sklearn.metrics import mean_absolute_percentage_error, r2_score

# --- 1. Bersihkan dan siapkan data
df = df.dropna(subset=['Description', 'Rating', 'Category'])
df = df.drop_duplicates(subset=['Place_Id', 'Description']).reset_index(drop=True)

# --- 2. Membangun model embedding dari awal menggunakan LSTM dengan Regularisasi
class TextEmbeddingModel(tf.keras.Model):
    def __init__(self, vocab_size, embedding_dim, sequence_length):
        super(TextEmbeddingModel, self).__init__()
        self.embedding = tf.keras.layers.Embedding(vocab_size, embedding_dim, input_length=sequence_length)
        self.lstm = tf.keras.layers.LSTM(64, return_sequences=False, kernel_regularizer=tf.keras.regularizers.l2(0.01))  # Regularisasi L2
        self.dropout = tf.keras.layers.Dropout(0.2)  # Dropout untuk mencegah overfitting
        self.dense = tf.keras.layers.Dense(128, activation='relu')

    def call(self, inputs):
        x = self.embedding(inputs)
        x = self.lstm(x)
        x = self.dropout(x)
        return self.dense(x)

# --- 3. Tokenisasi dan padding deskripsi teks
tokenizer = tf.keras.preprocessing.text.Tokenizer(num_words=10000)
tokenizer.fit_on_texts(df['Description'])
sequences = tokenizer.texts_to_sequences(df['Description'])
X = tf.keras.preprocessing.sequence.pad_sequences(sequences, maxlen=100)

# --- 4. Membuat model dan melatihnya (dengan validasi dan early stopping)
model = TextEmbeddingModel(vocab_size=10000, embedding_dim=128, sequence_length=100)
model.compile(loss='mse', optimizer='adam', metrics=['mae', metrics.RootMeanSquaredError()])  # Menambahkan MAE dan RMSE

# Pisahkan data untuk pelatihan dan validasi (80% untuk pelatihan, 20% untuk validasi)
y = np.random.random(size=(len(df), 128))  # Label acak (embedding output)

# Gunakan validation_split untuk pembagian otomatis data validasi
early_stopping = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)

history = model.fit(X, y, epochs=50, batch_size=32, validation_split=0.2, callbacks=[early_stopping])

# --- 5. Fungsi untuk mendapatkan representasi vektor deskripsi menggunakan model yang dilatih
def get_description_embedding(description):
    seq = tokenizer.texts_to_sequences([description])
    padded = tf.keras.preprocessing.sequence.pad_sequences(seq, maxlen=100)
    return model.predict(padded)[0]

# --- 6. Membuat vektor embedding untuk setiap deskripsi tempat wisata
df['embedding'] = df['Description'].apply(get_description_embedding)

# --- 7. Fungsi untuk menghitung cosine similarity antar tempat menggunakan TensorFlow
def cosine_similarity_tf(embedding1, embedding2):
    norm_embedding1 = tf.linalg.l2_normalize(embedding1)
    norm_embedding2 = tf.linalg.l2_normalize(embedding2)
    return tf.reduce_sum(norm_embedding1 * norm_embedding2)

# --- 8. Hitung cosine similarity antara semua tempat
def calculate_similarity_matrix(df):
    embeddings = np.array(df['embedding'].tolist())
    num_places = len(embeddings)
    similarity_matrix = np.zeros((num_places, num_places))

    for i in range(num_places):
        for j in range(num_places):
            similarity_matrix[i, j] = cosine_similarity_tf(
                tf.convert_to_tensor(embeddings[i]), tf.convert_to_tensor(embeddings[j])
            ).numpy()

    return similarity_matrix

# --- 9. Hitung similarity matrix
cosine_sim = calculate_similarity_matrix(df)

# --- 10. Fungsi rekomendasi berdasarkan input kategori
def rekomendasi_top10_kategori(kategori_input):
    # Filter data sesuai kategori
    df_kategori = df[df['Category'].str.lower() == kategori_input.lower()].copy()

    if df_kategori.empty:
        return f"❌ Tidak ditemukan tempat wisata dengan kategori: '{kategori_input}'"

    # Ambil index relatif (berdasarkan posisi di df yang sudah reset_index)
    idx_kategori = df_kategori.index.tolist()

    # Subset similarity matrix
    similarity_scores = cosine_sim[idx_kategori][:, idx_kategori]

    # Hitung rata-rata similarity dari tiap tempat ke tempat lain
    avg_similarity = similarity_scores.mean(axis=1)

    # Tambahkan skor similarity ke dataframe kategori
    df_kategori['avg_similarity'] = avg_similarity

    # Urutkan berdasarkan rating tertinggi dan similarity
    df_top = df_kategori.sort_values(by=['Rating', 'avg_similarity'], ascending=False)

    # Tampilkan 10 tempat wisata teratas
    return df_top[['Place_Name', 'Category', 'City_x', 'Rating']].head(10)

# --- 11. Contoh penggunaan
kategori_user = "Cagar Alam"  # Input user
output_top10 = rekomendasi_top10_kategori(kategori_user)

print(f"🎯 Top 10 tempat wisata berdasarkan rating tertinggi dalam kategori: '{kategori_user}'\n")
print(output_top10)
```

# Menambahkan fungsi R² dan MAPE untuk dicetak dan divisualisasikan

```python
def print_r2_mape(history, X_test, y_test):
    # Prediksi pada data uji
    y_pred = model.predict(X_test)

    # Menghitung R² (Koefisien Determinasi)
    r2 = r2_score(y_test, y_pred)
    print(f"R² (Training): {r2:.4f}")

    # Menghitung MAPE (Mean Absolute Percentage Error)
    mape = mean_absolute_percentage_error(y_test, y_pred)
    print(f"MAPE (Training): {mape*100:.2f}%")
```

# Visualisasi R² dan MAPE selama pelatihan

```python
def plot_r2_mape(history):
    epochs = range(1, len(history.history['mae']) + 1)

    # Plot R²
    plt.figure(figsize=(10, 6))
    plt.plot(epochs, history.history['mae'], label='Training MAE', color='blue')
    plt.plot(epochs, history.history['val_mae'], label='Validation MAE', color='orange')
    plt.title('R² vs Epochs')
    plt.xlabel('Epoch')
    plt.ylabel('R²')
    plt.legend()
    plt.show()

    # Plot MAPE
    plt.figure(figsize=(10, 6))
    plt.plot(epochs, history.history['mae'], label='Training MAPE', color='blue')
    plt.plot(epochs, history.history['val_mae'], label='Validation MAPE', color='orange')
    plt.title('MAPE vs Epochs')
    plt.xlabel('Epoch')
    plt.ylabel('MAPE')
    plt.legend()
    plt.show()

# --- 14. Panggil fungsi untuk visualisasi R² dan MAPE
plot_r2_mape(history)
```

# Visualisasi MAE dan RMSE
